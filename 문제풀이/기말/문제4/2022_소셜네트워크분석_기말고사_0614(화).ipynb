## 문제 1

```python
# dat1에서 결측치가 200 이상인 column을 제거하시오 (또는 결측치가 200미만인 column 만 가져오시오)
# (dat1은 채점시 변경 가능)

#=== 코드작성 ===# 
dat1.head()
sr1 = np.sum(dat1.isnull(), axis=0)
col = sr1[sr1>=200].index[0]
df_p1 = dat1.drop(columns=col)
#=== 코드작성 끝===# 
print('### df_p1: dat1에서 결측치가 200 이상인 column제거한 dataframe')
df_p1
```

## 문제 2

```python
# dat1에서 결측치가 제일 많은 컬럼 2개를 제거하시오
# (dat1은 채점시 변경 가능)

#=== 코드작성 ===# 
cols = list(np.sum(dat1.isnull(), axis=0).sort_values(ascending=False).iloc[:2].index)
df_p2 = dat1.drop(columns=cols)
#=== 코드작성 끝===# 
print('### df_p2: dat1에서 결측치가 제일 많은 컬럼 2개를 제거한 dataframe')
df_p2
```

```python
# dat1에서 C1컬럼이 'A'이고 C4 컬럼에 '바다'가 포함되어있는 행들에 대하여,
# C6 컬럼의 평균값을 구하시오

#=== 코드작성 ===#
df3 = dat1.loc[(dat1['C1']=='A')&(dat1['C4'].str.contains('바다'))]
m3 = df3['C6'].mean()
#=== 코드작성 끝===# 

print("### m3: dat1에서 C1컬럼이 'A'이고 C4 컬럼에 '바다'가 포함되어있는 행들에 대하여, C6 컬럼의 평균값")
m3
```

## 문제 5

```python
# 부산시의 상점들에 대하여,
# '상권업종대분류명' 별, '시군구명' 별 "상점수의 bar graph"를
# 'seaborn countplot'을 사용하여 아래그림과 같이 나타내시오

plt.figure(figsize=(20, 4), dpi=100)

#=== 코드작성 ===#
busan = dat2.loc[dat2['시도명']=='부산광역시']

_=sns.countplot(data=busan,
               x='상권업종대분류명', hue='시군구명')

#=== 코드작성 끝===# 
```

## 문제 6

```python
# 부산시의 '상권업종대분류명'이 '의료'인 상점들에 대하여,
# '시군구명' 별, '상권업종중분류명' 별 "상점수의 bar graph"를
# 'pandas plot bar'를 사용하여 아래그림과 같이 나타내시오

fig=plt.figure(figsize=(20, 4), dpi=100)
ax1=fig.subplots()

#=== 코드작성 ===#
df6 = busan.loc[busan['상권업종대분류명']=='의료']
df66 = df6[['시군구명','상권업종중분류명']].value_counts().unstack()
df66
_=df66.plot.bar(ax=ax1)
#=== 코드작성 끝===# 
```

## 문제 7

```python
# 부산시의 상점에 대하여,
# '시군구명' 별, '상권업종대분류명' 상점수의 bar graph"를
# 'seaborn catplot'를 사용하여 아래그림과 같이 나타내시오
# (단, catplot parameter: height=3, aspect=2)

#=== 코드작성 ===#
df7 = busan.groupby(['시군구명','상권업종대분류명'])['상호명'].count().reset_index()
df7.columns = ['시군구명','상권업종대분류명','상호수']
df7.head()

g = sns.catplot(data=df7,
                x = '시군구명', y='상호수',
                kind='bar',
                col='상권업종대분류명', col_wrap=2,
                height=3, aspect=2)

#=== 코드작성 끝===# 

for ax in g.axes.flat:
    _=ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize=12)
    _=ax.set_title(ax.get_title(), fontsize=12)
    
plt.tight_layout()
```

## 문제 9

```python
# 문제8과 같은 그림을 folium을 활용하여 실제 지도위에 나타내시오
# (icon color는 아래의 gu_color_dic에 있는 정보를 활용하시오)
# (marker는 tooltip 스타일로 하고, tooltip 문구는 상호명으로 하시오)

gu_color_dic={'해운대구': 'blue','사상구': 'red', '부산진구': 'orange', '사하구': 'green'}

#=== 코드작성 ===#
lat = df8['위도'].mean()
long = df8['경도'].mean()

m = folium.Map([lat, long], zoom_start=13, tiles = 'Stamen Toner')

df9 = df8.loc[dat2['시군구명'].isin(['해운대구','사상구', '부산진구', '사하구'])]

for i in df9.index:
    t1 = df8.loc[i,'상호명']
    lat = df8.loc[i,'위도']
    long = df8.loc[i,'경도']
    gu = df8.loc[i,'시군구명']
    _=folium.Marker([lat,long], tooltip=t1, radius=3, 
                    icon = folium.Icon(color=gu_color_dic[gu])).add_to(m)

m
#=== 코드작성 끝===# 
```
