## 문제 6

```python
# 부산시의 '상권업종대분류명'이 '의료'인 상점들에 대하여,
# '시군구명' 별, '상권업종중분류명' 별 "상점수의 bar graph"를
# 'pandas plot bar'를 사용하여 아래그림과 같이 나타내시오

fig=plt.figure(figsize=(20, 4), dpi=100)
ax1=fig.subplots()

#=== 코드작성 ===#
df_p6=dat2.loc[(dat2['시도명']=='부산광역시') & (dat2['상권업종대분류명']=='의료')]
p6_gp=df_p6.groupby(["시군구명", "상권업종중분류명"])["상호명"].count().unstack()
p6_gp.plot.bar(ax=ax1)
#=== 코드작성 끝===# 
```

## 문제 7

```python
# 부산시의 상점에 대하여,
# '시군구명' 별, '상권업종대분류명' 상점수의 bar graph"를
# 'seaborn catplot'를 사용하여 아래그림과 같이 나타내시오
# (단, catplot parameter: height=3, aspect=2)

#=== 코드작성 ===#
df_p7=dat2.loc[(dat2['시도명']=='부산광역시')]

df_p7=df_p7.groupby(["시군구명", "상권업종대분류명"])["상호명"].count().reset_index()
df_p7 = df_p7.rename(columns={"상호명":"상호수"})
g=sns.catplot(data=df_p7, x="시군구명", y="상호수", kind='bar', col='상권업종대분류명',col_wrap=2, height=3, aspect=2)
#=== 코드작성 끝===# 

for ax in g.axes.flat:
    _=ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize=12)
    _=ax.set_title(ax.get_title(), fontsize=12)
    
plt.tight_layout()
```

## 문제 9

```python
# 문제8과 같은 그림을 folium을 활용하여 실제 지도위에 나타내시오
# (icon color는 아래의 gu_color_dic에 있는 정보를 활용하시오)
# (marker는 tooltip 스타일로 하고, tooltip 문구는 상호명으로 하시오)

gu_color_dic={'해운대구': 'blue','사상구': 'red', '부산진구': 'orange', '사하구': 'green'}

#=== 코드작성 ===#
df_p9=dat2.loc[(dat2['시도명']=='부산광역시') & (dat2['상권업종대분류명']=='숙박')]
lat = df_p9["위도"].mean()
long = df_p9["경도"].mean()

m = folium.Map([lat, long], zoom_start=12, tiles="stamen toner")

for i in df_p9.index:
    sub_lat = df_p9.loc[i, "위도"]
    sub_long = df_p9.loc[i, "경도"]
    title = df_p9.loc[i, "상호명"]
    gu = df_p9.loc[i, "시군구명"]

    _=folium.Marker(
        [sub_lat,sub_long], 
        icon=folium.Icon(color=gu_color_dic[gu]),
        tooltip=title).add_to(m)
m
#=== 코드작성 끝===# 
```

```python
dat3 = pd.read_csv("data/전국도시공원표준데이터.csv", encoding="cp949")
dat3 = dat3.loc[(dat3["경도"] < 132) & (dat3["위도"] > 32)].copy()

## 시도컬럼 생성
dat3["시도"] = dat3["소재지지번주소"].str.split(expand=True)[0]

## 경기도, 서울특별시만 선택
dat3=dat3.loc[dat3['시도'].isin(['경기도','서울특별시'])][['공원명','공원구분','공원면적','지정고시일','시도']]
dat3=dat3.dropna()
dat3
```

## 문제 10 (풀이1)

```python
# 2007년에 경기도와 서울특별시에 신설된 공원들에 대하여,
# 월별 누적 신설 공원수를 pandas plot을 활용하여 bar graph로 나타내시오

df_p10=dat3.copy()
fig=plt.figure(figsize=(10, 8), dpi=100)
ax1=fig.subplots()

#=== 코드작성 ===#
df_p10.head(1)

df_p10['지정고시일']=pd.to_datetime(df_p10['지정고시일'])
df_p10.head(1)

df_p10['연도']=df_p10['지정고시일'].dt.year
df_p10.head(1)

df_p10['월']=df_p10['지정고시일'].dt.month

df_p10=df_p10.loc[df_p10['연도']==2007]

# aa
df10_s=df_p10.loc[df_p10['시도']=='서울특별시']
df10_k=df_p10.loc[df_p10['시도']=='경기도']

s_count=df10_s['월'].value_counts()
k_count=df10_k['월'].value_counts()

sk_count=pd.concat([k_count, s_count], axis=1).sort_index()
sk_count.columns=['경기도', '서울특별시']
# bb

## cumsum 함수
sk_count_cumsum=sk_count.cumsum()

# pandas plot
sk_count_cumsum.iloc[::-1].plot.barh(ax=ax1)

#=== 코드작성 끝===# 
```

## 문제 10 (풀이2)

```python
# 2007년에 경기도와 서울특별시에 신설된 공원들에 대하여,
# 월별 누적 신설 공원수를 pandas plot을 활용하여 bar graph로 나타내시오

df_p10=dat3.copy()
fig=plt.figure(figsize=(10, 8), dpi=100)
ax1=fig.subplots()

#=== 코드작성 ===#
df_p10['지정고시일']=pd.to_datetime(df_p10['지정고시일'])
df_p10['연도']=df_p10['지정고시일'].dt.year
df_p10['월']=df_p10['지정고시일'].dt.month

df_p10=df_p10.loc[df_p10['연도']==2007]

# (풀이1)에서 aa~bb 까지의 코드를 실행하는 한 줄 코드
sk_count=df_p10.groupby(['월','시도'])['공원명'].count().unstack() 

## cumsum 함수
sk_count_cumsum=sk_count.cumsum()

# pandas plot
sk_count_cumsum.iloc[::-1].plot.barh(ax=ax1)

#=== 코드작성 끝===# 
```

```python
dat4 = pd.read_csv("data/전국도시공원표준데이터.csv", encoding="cp949")
dat4 = dat4.loc[(dat4["경도"] < 132) & (dat4["위도"] > 32)].copy()

## 시도컬럼 생성
dat4["시도"] = dat4["소재지지번주소"].str.split(expand=True)[0]
dat4["시도"] = dat4["시도"].replace("강원", "강원도")
dat4=dat4[['공원명','공원구분','공원보유시설(운동시설)','시도']].copy()
dat4
```

## 문제 11

```python
# 각 시도 별, 공원보유시설(운동시설)에 1) 축구장 2) 배드민턴장 3) 농구장 포함 공원의 개수를 bar graph로 그리시오
# (단, 농구장이 가장 많은 순서로 시도를 sorting하여 나타내시오)
# (pandas plot bar 또는 seaborn barplot 모두 상관 없음)

fig=plt.figure(figsize=(20,5))
ax1=fig.subplots()

#=== 코드작성 ===#
soc=dat4.loc[dat4['공원보유시설(운동시설)'].str.contains('축구장', na=False)]['시도'].value_counts()
bed=dat4.loc[dat4['공원보유시설(운동시설)'].str.contains('배드민턴장', na=False),'시도'].value_counts()
bas=dat4.loc[dat4['공원보유시설(운동시설)'].str.contains('농구장', na=False),'시도'].value_counts()

df_con=pd.concat([soc, bed,bas], axis=1).fillna(0)
df_con.columns=['축구장','배드민턴장','농구장']
df_con=df_con.sort_values(by='농구장', ascending=False)
df_con.plot.bar(ax=ax1)
#=== 코드작성 끝===# 
```
