## 문제1

```python
# 주어진 df1의 column 'col_a'가
# 3자리 숫자인 경우에는 그 숫자를, 그렇지 않은 경우는 nan을 넣은 column 'col_b'를 만드는 코드이다
# 함수 f1을 구현하시오

# * 그렇지 않은 경우
# 1) 3자리가 아닌 경우
# 2) 영문자가 포함되어있는 경우 (*힌트: isdigit())
# 

## 데이터 df1 생성
df1=pd.DataFrame({'col1':np.random.randint(10,500,size=30),
                  'col2':np.random.choice(['']*10+['a','b','c'], size=30)})

df1['col1']=df1['col1'].astype(str)
df1['col_a'] = df1['col1']+df1['col2']
df1.drop(['col1','col2'], axis='columns', inplace=True)
## df1 생성 끝

## 코드작성
def func1(x):
    if x!=x:
        return np.nan
    elif x.isdigit() and len(x)==3:
        return x
    else:
        return np.nan
    
df1['col_b'] = df1['col_a'].map(func1)

df1
```

## 문제 2, 3, 4

### 데이터 생성

```python
df1=pd.DataFrame({'랭크':np.random.choice(['R1','R2','R3'], size=1000),
                  '학년':np.random.randint(1,5,size=1000),
                  '그룹':np.random.choice(list('ABCD'), size=1000)})

def jap_func(x):
    dic_jap={'R1':10, 'R2':20, 'R3':30}
    return np.random.normal(dic_jap[x],1)

def mat_func(x):
    return np.random.normal(x*2,1)

def eng_func(x):
    return np.random.normal(10,dic_eng[x])


df1['일어']=df1['랭크'].map(jap_func)
df1['수학']=df1['학년'].map(mat_func)

shf_gp=np.random.choice(list('ABCD'), size=4, replace=False)
dic_eng=dict(zip(shf_gp, [1,3,5,7]))
df1['영어']=df1['그룹'].map(eng_func)

df1.sample(10)
```

### 문제2

```python
# 주어진 df1을 가지고, 랭크 별 일어 평균점수를 bar graph로 나타내는데,
# ax1) pandas의 plot.bar 함수 사용
# ax2) seaborn의 barplot 함수 사용
# (ax2의 x tick label 순서는 관계 없음)

fig=plt.figure(figsize=(10,3), dpi=100)
(ax1,ax2)=fig.subplots(1,2)

## 코드작성
dfm = df1.groupby(['랭크'])['일어'].mean().to_frame()
_=dfm.plot.bar(ax=ax1)

_=sns.barplot(data=df1, x='랭크', y='일어', estimator=np.mean, ax=ax2)

_=ax1.set_title('pandas plot.bar')
_=ax2.set_title('seaborn barplot')
```

### 문제3

```python
# 주어진 df1을 가지고, 각 그룹 별 영어 점수를 seaborn으로 ax1, ax2에 box plot 하는데
# ax1) x축 정렬(sort) 없이 plot .
# ax2) x축의 순서를 standard deviation 값으로 (사용 함수: np.std) 으로 정렬(sort) 후 plot.
# (ax2의 경우 데이터가 랜덤하게 생성되기 때문에, 특정 그룹의 order로 지정하는 것은 안됨)

fig=plt.figure(figsize=(10,4), dpi=100)
(ax1,ax2)=fig.subplots(1,2)

## 코드작성
_=sns.boxplot(data=df1, 
              x='그룹', y='영어',
              ax=ax1)

dfs = df1.groupby(['그룹'])['영어'].std().sort_values()

_=sns.boxplot(data=df1, 
              x='그룹', y='영어',
              order=dfs.index,
              ax=ax2)
```

### 문제4

```python
# 주어진 df1에서에서, (학년, 랭크) 별 일어점수의 최대값을 아래와 같이 bar 그래프로 그리시오
# ax1) pandas plot 사용하여 그림.
# ax2) seaborn barplot 사용하여 그림.

# (평균이 아니라 최대값에 주의. np.max 함수 사용, aggfunc, estimator parameter 사용)
# (평균으로 그리면 감점 있음)
# (ax1의 x축 tick rotation은 pandas plot 함수 내 에서, 'rot' parameter 사용)

fig=plt.figure(figsize=(10,4), dpi=100)
(ax1,ax2)=fig.subplots(1,2)

## 코드작성

dfmax = df1.pivot_table(index='학년',columns='랭크',values='일어',aggfunc=np.max)
dfmax

_=dfmax.plot.bar(rot=0, ax=ax1)
_=ax1.legend(bbox_to_anchor=(1.02, 1), loc=2, title='랭크')

_=sns.barplot(data=df1, x='학년', y='일어', hue='랭크', estimator=np.max, ci=0, ax=ax2)
_=ax2.legend(bbox_to_anchor=(1.02, 1), loc=2, title='랭크')


fig.tight_layout()
```

## 문제 5,6

### 데이터 생성

```python
df_last = pd.read_csv("전국평균 분양가격 (2015년10월~2019년12월).csv", encoding="cp949")
df_last["분양가격"] = pd.to_numeric(df_last["분양가격(㎡)"], errors='coerce')
df_last["평당분양가격"] = df_last["분양가격"] * 3.3
df_last["전용면적"] = df_last["규모구분"].str.replace("전용면적", "")
df_last["전용면적"] = df_last["전용면적"].str.replace("초과", "~")
df_last["전용면적"] = df_last["전용면적"].str.replace("이하", "")
df_last["전용면적"] = df_last["전용면적"].str.replace(" ", "")
df_last = df_last.drop(["규모구분", "분양가격(㎡)"], axis=1)
df_last.head()
```

### 문제 5

```python
# 주어진 df_last에서,
# ax1) 지역 별 평당분양가격을 boxplot으로 그리시오 (pandas boxplot 함수 사용)
# ax2) 지역 별, 월 별 평당분양가격의 평균을 담은 dataframe을 g1과 같이 만들고,
#       g1을 가지고 지역 별로 boxplot을 그리시오 (pandas plot.box 함수 사용)

fig=plt.figure(figsize=(10,6), dpi=100)
(ax1,ax2)=fig.subplots(2,1).flatten()

## 코드작성
_=df_last.boxplot(column='평당분양가격', by='지역명',ax=ax1)

print('##### g1 #####')
dfm = df_last.pivot_table(index='월',columns='지역명',values='평당분양가격',aggfunc=np.mean)
dfm
_=dfm.plot.box(ax=ax2)
fig.tight_layout()
```
