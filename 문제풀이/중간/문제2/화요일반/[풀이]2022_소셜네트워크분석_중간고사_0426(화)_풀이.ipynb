```python
# C4컬럼의 각 값에 3을 곱하여 C4_a 컬럼에 넣으시오
# 단, 괄호에 영어로 표시가 되어있는 값들은 괄호와 함께 영어표시를 제거하고 3을 곱하시오

## 코드작성
def func1(x):
    return int(x.split("(")[0])*3

df_p1['C4_a']=df_p1['C4'].map(func1)
## 코드작성 끝

df_p1
```

```python
df_last = pd.read_csv("data/전국평균 분양가격 (2015년10월~2019년12월).csv", encoding="cp949")
df_last["분양가격"] = pd.to_numeric(df_last["분양가격(㎡)"], errors='coerce')
df_last["평당분양가격"] = df_last["분양가격"] * 3.3
df_last["전용면적"] = df_last["규모구분"].str.replace("전용면적", "")
df_last["전용면적"] = df_last["전용면적"].str.replace("초과", "~")
df_last["전용면적"] = df_last["전용면적"].str.replace("이하", "")
df_last["전용면적"] = df_last["전용면적"].str.replace(" ", "")
df_last = df_last.drop(["규모구분", "분양가격(㎡)"], axis=1)
df_last
```

## 문제 6

```python
# pivot_table을 사용하여 월별, 연도별 평당분양가격의 "중앙값"을 아래와 같이 나타내시오 (단, unstack 사용 불가)

## 코드작성
df7 = pd.pivot_table(df_last, index="월", columns= "연도", values="평당분양가격", aggfunc=np.median)
## 코드작성 끝

df7
```

## 문제 7

```python
# 연도별, 전용면적별 평당분양가격의 "평균"을 구한뒤, 결과를 출력하고,
# 그 결과 dataframe을 주어진 ax1에 각 연도별로 boxplot을 그리시오 (하나의 box는 전용면적종류(5개) 만큼의 점으로 구성)

fig=plt.figure(figsize=(6,3), dpi=100)
ax1=fig.subplots()

## 코드작성
p = df_last.pivot_table(index="전용면적", columns="연도", values="평당분양가격")
p
p.plot.box(ax=ax1)
## 코드작성 끝
```

## 문제 8

```python
from matplotlib.ticker import MaxNLocator
# 주어진 ax1에,
# pandas plot을 사용하여 전용면적 별, 연도 별 평당분양가격의 "최소값"을 아래와 같이 line graph로 나타내시오
# 그리고, x축의 tick을 정수로 표현하시오

fig=plt.figure(figsize=(6,3), dpi=100)
ax1=fig.subplots()

## 코드작성
p = df_last.pivot_table(index="연도", columns="전용면적", values="평당분양가격", aggfunc=np.min)
p.plot(ax=ax1)
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
## 코드작성 끝

ax1.legend(bbox_to_anchor=(1.05, 1), loc=2)

```

## 문제 9

```python
# 주어진 ax1에,
# seaborn plot을 사용하여 전용면적 별, 연도별, 평당분양가격의 "중앙값"을 아래와 같이 bar graph로 나타내시오
# 그리고 conficence interval을 100으로 하시오

fig=plt.figure(figsize=(10,4), dpi=100)
ax1=fig.subplots()

## 코드작성
sns.barplot(data=df_last, x="전용면적", y="평당분양가격", hue='연도', ci=100, estimator=np.median)
## 코드작성 끝
ax1.legend(bbox_to_anchor=(1.05, 1), loc=2)
```

## 문제 10

```python
df_p10=pd.DataFrame()
df_p10['col1']=(pd.Series(np.random.randint(1,20,size=10)).astype(str)+'_'+
               pd.Series(np.random.choice(['a','b','c','d','e'], size=10))+'_'+
               pd.Series(np.random.randint(1,20,size=10)).astype(str)+'_'+
               pd.Series(np.random.choice(['a','b','c','d','e'], size=10)))
df_p10

# 주어진 df_p10의 col1에는 숫자_알파벳_숫자_알파벳 이런 식으로 데이터가 저장되어있다.
# col2 컬럼을 만들어서 숫자들의 합을 넣으시오

## 코드작성
def func1(x):
    n1=int(x.split('_')[0])
    n2=int(x.split('_')[2])
    return n1+n2
df_p10['col2']=df_p10['col1'].map(func1)
## 코드작성 끝

df_p10
```
